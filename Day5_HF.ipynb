```python
import requests

HF_TOKEN = os.environ["HF_TOKEN"]

def hf_zero_shot_api(
    text,
    labels
):

    url = (
        "https://router.huggingface.co/"
        "hf-inference/models/"
        "facebook/bart-large-mnli"
    )

    headers = {
        "Authorization":
        f"Bearer {HF_TOKEN}"
    }

    payload = {
        "inputs":
        text,

        "parameters":{
            "candidate_labels":
            labels
        }
    }

    r = requests.post(
        url,
        headers=headers,
        json=payload
    )

    return r.json()
```

```python
from transformers import pipeline

sentiment = pipeline(
    "sentiment-analysis",
    model=
    "distilbert-base-uncased-finetuned-sst-2-english"
)

answers = [

"I really enjoyed working on the team.",

"Everyone else was slow.",

"I learned a lot from my mentor.",

"I had to redo teammate work.",

"My internship was great."
]

for a in answers:

    result = sentiment(a)[0]

    print(
        result["label"],
        round(
            result["score"],
            2
        )
    )
```

```python
import time

def time_fn(fn):

    arr=[]

    for _ in range(3):

        s=time.time()

        fn()

        arr.append(
            time.time()-s
        )

    return min(arr),sum(arr)/3


def local():

    classifier(

        "React dashboard",

        candidate_labels=[
            "frontend",
            "backend"
        ]
    )

mn,av=time_fn(local)

print(
"LOCAL"
)

print(
"min:",
round(mn,2)
)

print(
"avg:",
round(av,2)
)
```
